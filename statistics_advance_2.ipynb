1  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    standard_error = population_std / np.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / standard_error
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    decision = "Reject" if p_value < alpha else "Fail to Reject"

    print(f"Z-score: {z_score:.3f}")
    print(f"P-value: {p_value:.3f}")
    print(f"Decision: {decision} the null hypothesis")

# Example data
sample = np.random.normal(50, 10, 30)  # Simulating a sample
population_mean = 52
population_std = 10

z_test(sample, population_mean, population_std)
```

9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence_level=0.95):
    # Calculate sample mean, sample standard deviation, and sample size
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # ddof=1 for sample standard deviation
    sample_size = len(sample)

    # Find the critical t-value (using t-distribution for small sample sizes)
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)

    # Standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate the margin of error
    margin_of_error = t_critical * standard_error

    # Confidence Interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
np.random.seed(0)  # Set seed for reproducibility
sample = np.random.normal(50, 10, 30)  # Sample of 30 with mean=50 and std=10

# Calculate 95% confidence interval for the sample mean
lower, upper = confidence_interval(sample, confidence_level=0.95)

print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

10 Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence_level=0.95):
    # Calculate sample mean, sample standard deviation, and sample size
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # ddof=1 for sample standard deviation
    sample_size = len(sample)

    # Calculate alpha (significance level)
    alpha = 1 - confidence_level

    # Find the critical value from the t-distribution (or Z-distribution for large samples)
    t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)

    # Standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate margin of error
    margin_of_error = t_critical * standard_error

    return margin_of_error

# Example usage:
np.random.seed(0)  # Set seed for reproducibility
sample = np.random.normal(50, 10, 30)  # Sample of 30 with mean=50 and std=10

# Calculate margin of error for 95% confidence level
margin_error = margin_of_error(sample, confidence_level=0.95)

print(f"Margin of Error: {margin_error:.2f}")
```

11  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayesian_inference(prior, likelihood, evidence):
    """
    Perform Bayesian Inference using Bayes' Theorem.

    Parameters:
    - prior: The prior probability of the hypothesis (P(H))
    - likelihood: The likelihood of observing the evidence given the hypothesis (P(E|H))
    - evidence: The overall probability of observing the evidence (P(E))

    Returns:
    - posterior: The updated probability of the hypothesis given the evidence (P(H|E))
    """
    # Apply Bayes' Theorem: P(H|E) = (P(E|H) * P(H)) / P(E)
    posterior = (likelihood * prior) / evidence
    return posterior

# Example data for a diagnostic test:

# 1. Prior probability (P(H)): The probability of having the disease (before test)
prior = 0.01  # 1% of people have the disease

# 2. Likelihood (P(E|H)): The probability of a positive test given the person has the disease
likelihood = 0.95  # 95% chance of testing positive if you have the disease

# 3. Evidence (P(E)): The total probability of testing positive (whether or not you have the disease)
# This includes both true positives and false positives:
# P(E) = P(E|H) * P(H) + P(E|~H) * P(~H)
# P(E|~H) = False positive rate, and P(~H) = 1 - P(H)
false_positive_rate = 0.05  # 5% chance of testing positive if you don't have the disease
evidence = (likelihood * prior) + (false_positive_rate * (1 - prior))

# Perform Bayesian Inference
posterior = bayesian_inference(prior, likelihood, evidence)

# Output the result
print(f"The probability that the person has the disease given a positive test result is {posterior:.4f}")
```

13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate the expected frequencies for a Chi-Square test given observed data.

    Parameters:
    - observed: A 2D numpy array or list representing the observed contingency table.

    Returns:
    - expected: The expected frequencies based on the observed data.
    """
    # Convert the observed data into a numpy array (if it's not already)
    observed = np.array(observed)

    # Calculate the row totals and column totals
    row_totals = observed.sum(axis=1, keepdims=True)  # Sum across rows (axis=1)
    col_totals = observed.sum(axis=0, keepdims=True)  # Sum across columns (axis=0)

    # Calculate the total number of observations
    total = observed.sum()

    # Calculate expected frequencies for each cell
    expected = (row_totals * col_totals) / total

    return expected

# Example observed data: Contingency table for two categorical variables
observed_data = [
    [30, 10],  # 30 Male prefer Product A, 10 Male prefer Product B
    [20, 25]   # 20 Female prefer Product A, 25 Female prefer Product B
]

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Output the results
print("Observed Frequencies:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(expected_frequencies)
```

14  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

```python
import numpy as np
import scipy.stats as stats

def chi_square_goodness_of_fit(observed, expected):
    """
    Perform a Chi-Square Goodness-of-Fit test to compare observed data to expected distribution.

    Parameters:
    - observed: A list or numpy array of observed frequencies.
    - expected: A list or numpy array of expected frequencies.

    Returns:
    - chi2_stat: The chi-square test statistic.
    - p_value: The p-value of the test.
    """
    # Convert observed and expected data to numpy arrays for element-wise operations
    observed = np.array(observed)
    expected = np.array(expected)

    # Calculate the chi-square statistic
    chi2_stat = np.sum((observed - expected) ** 2 / expected)

    # Degrees of freedom: number of categories - 1
    dof = len(observed) - 1

    # Calculate the p-value based on the chi-square distribution
    p_value = 1 - stats.chi2.cdf(chi2_stat, dof)

    return chi2_stat, p_value

# Example data: Observed frequencies (e.g., observed dice rolls) vs expected distribution (uniform distribution)
observed_data = [5, 6, 4, 9, 8, 3]  # Observed frequencies for 6 categories (dice rolls)
expected_data = [7, 7, 7, 7, 7, 7]  # Expected frequencies for a fair die (uniform distribution)

# Perform Chi-Square Goodness-of-Fit test
chi2_stat, p_value = chi_square_goodness_of_fit(observed_data, expected_data)

# Output the results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The observed data does not follow the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data follows the expected distribution.")
```

```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2):
    """
    Perform an F-test to compare the variances of two random samples.

    Parameters:
    - sample1: First sample data (list or numpy array).
    - sample2: Second sample data (list or numpy array).

    Returns:
    - f_statistic: The F-test statistic.
    - p_value: The p-value of the test.
    """
    # Calculate the variances of the two samples
    var1 = np.var(sample1, ddof=1)  # Sample variance (ddof=1 for unbiased estimate)
    var2 = np.var(sample2, ddof=1)

    # Ensure that the larger variance is in the numerator for the F-test
    if var1 > var2:
        f_statistic = var1 / var2
        df1, df2 = len(sample1) - 1, len(sample2) - 1
    else:
        f_statistic = var2 / var1
        df1, df2 = len(sample2) - 1, len(sample1) - 1

    # Calculate the p-value for the F-distribution
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

    return f_statistic, p_value

# Example: Simulate two random samples and perform the F-test
np.random.seed(42)

# Simulate two random samples from normal distributions with different variances
sample1 = np.random.normal(loc=0, scale=2, size=50)  # Mean=0, Std=2, n=50
sample2 = np.random.normal(loc=0, scale=4, size=50)  # Mean=0, Std=4, n=50

# Perform the F-test
f_statistic, p_value = f_test(sample1, sample2)

# Output the results
print(f"F-test Statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The variances are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The variances are not significantly different.")
```

```python
import numpy as np
import scipy.stats as stats

def perform_anova(*groups):
    """
    Perform an ANOVA test to compare the means of multiple groups.

    Parameters:
    - *groups: Multiple lists or arrays of sample data (one for each group).

    Returns:
    - f_statistic: The F-test statistic.
    - p_value: The p-value of the test.
    """
    # Perform the ANOVA test
    f_statistic, p_value = stats.f_oneway(*groups)

    return f_statistic, p_value

# Example: Simulate data for three groups
np.random.seed(42)

# Group 1: Heights in cm for group 1
group1 = np.random.normal(loc=160, scale=10, size=30)  # Mean=160, Std=10, n=30

# Group 2: Heights in cm for group 2
group2 = np.random.normal(loc=170, scale=15, size=30)  # Mean=170, Std=15, n=30

# Group 3: Heights in cm for group 3
group3 = np.random.normal(loc=175, scale=12, size=30)  # Mean=175, Std=12, n=30

# Perform the ANOVA test
f_statistic, p_value = perform_anova(group1, group2, group3)

# Output the results
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: At least one group mean is significantly different.")
else:
    print("\nFail to reject the null hypothesis: The group means are not significantly different.")
```

19  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
import numpy as np
import pandas as pd
from scipy.stats import f_oneway, shapiro, levene
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(data, group_col, value_col):
    """
    Check the assumptions for ANOVA: normality, independence, and equal variance.

    Parameters:
        data (pd.DataFrame): The dataset containing the groups and their values.
        group_col (str): The column name representing the groups.
        value_col (str): The column name representing the values.

    Returns:
        dict: A dictionary with the results of the tests.
    """
    results = {}
    groups = data[group_col].unique()

    # Test for normality (Shapiro-Wilk test) within each group
    normality_results = {}
    for group in groups:
        stat, p = shapiro(data[data[group_col] == group][value_col])
        normality_results[group] = {'statistic': stat, 'p-value': p}
    results['normality'] = normality_results

    # Test for equal variance (Levene's test)
    group_values = [data[data[group_col] == group][value_col] for group in groups]
    stat, p = levene(*group_values)
    results['equal_variance'] = {'statistic': stat, 'p-value': p}

    # Independence is usually checked by study design (not programmatically), but we note it here
    results['independence'] = "Independence must be ensured by study design."

    return results

# Step 1: Generate sample data for three groups
group1 = np.random.normal(20, 5, 30)  # Mean=20, SD=5, N=30
group2 = np.random.normal(22, 5, 30)  # Mean=22, SD=5, N=30
group3 = np.random.normal(19, 5, 30)  # Mean=19, SD=5, N=30

# Combine data into a DataFrame
data = pd.DataFrame({
    'Score': np.concatenate([group1, group2, group3]),
    'Group': ['Group 1'] * len(group1) + ['Group 2'] * len(group2) + ['Group 3'] * len(group3)
})

# Step 2: Perform one-way ANOVA test
anova_result = f_oneway(group1, group2, group3)
print("ANOVA Results:")
print(f"F-statistic: {anova_result.statistic:.2f}")
print(f"p-value: {anova_result.pvalue:.4f}")

# Check ANOVA assumptions
assumptions = check_anova_assumptions(data, group_col='Group', value_col='Score')
print("\nAssumptions Check:")
for assumption, result in assumptions.items():
    print(f"{assumption.capitalize()}: {result}")

# Step 3: Visualize the results with a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Score', data=data, palette='Set2')
plt.title('Box Plot of Scores by Group', fontsize=16)
plt.xlabel('Group', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import f_oneway, shapiro, levene
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(data, group_col, value_col):
    """
    Check the assumptions for ANOVA: normality, independence, and equal variance.

    Parameters:
        data (pd.DataFrame): The dataset containing the groups and their values.
        group_col (str): The column name representing the groups.
        value_col (str): The column name representing the values.

    Returns:
        dict: A dictionary with the results of the tests.
    """
    results = {}
    groups = data[group_col].unique()

    # Test for normality (Shapiro-Wilk test) within each group
    normality_results = {}
    for group in groups:
        stat, p = shapiro(data[data[group_col] == group][value_col])
        normality_results[group] = {'statistic': stat, 'p-value': p}
    results['normality'] = normality_results

    # Test for equal variance (Levene's test)
    group_values = [data[data[group_col] == group][value_col] for group in groups]
    stat, p = levene(*group_values)
    results['equal_variance'] = {'statistic': stat, 'p-value': p}

    # Independence is usually checked by study design (not programmatically), but we note it here
    results['independence'] = "Independence must be ensured by study design."

    return results

# Step 1: Generate sample data for two factors and their interaction
np.random.seed(42)
A_levels = ["A1", "A2"]
B_levels = ["B1", "B2", "B3"]
data = []
for a in A_levels:
    for b in B_levels:
        group_scores = np.random.normal(loc=(20 + 2 * (A_levels.index(a)) + 3 * (B_levels.index(b))), scale=5, size=15)
        for score in group_scores:
            data.append([a, b, score])
data = pd.DataFrame(data, columns=["FactorA", "FactorB", "Score"])

# Step 2: Perform two-way ANOVA
two_way_model = ols('Score ~ C(FactorA) + C(FactorB) + C(FactorA):C(FactorB)', data=data).fit()
anova_table = sm.stats.anova_lm(two_way_model, typ=2)
print("Two-Way ANOVA Results:")
print(anova_table)

# Step 3: Visualize the results with interaction plots
plt.figure(figsize=(10, 6))
sns.pointplot(x="FactorB", y="Score", hue="FactorA", data=data, palette="Set1", dodge=True, markers=["o", "s"], linestyles=["-", "--"])
plt.title('Interaction Plot: FactorA and FactorB', fontsize=16)
plt.xlabel('Factor B', fontsize=14)
plt.ylabel('Mean Score', fontsize=14)
plt.legend(title="Factor A")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Step 1: Simulate data from a normal distribution
data = np.random.normal(20, 5, 30)  # Mean=20, SD=5, N=30

# Step 2: Set up the hypothesis test
# Null hypothesis: The population variance is equal to sigma^2 = 25 (SD = 5)
sigma_squared = 25
n = len(data)
sample_variance = np.var(data, ddof=1)
chi_square_stat = (n - 1) * sample_variance / sigma_squared

# Degrees of freedom
df = n - 1

# Critical values for a 95% confidence level (two-tailed test)
alpha = 0.05
chi2_critical_lower = chi2.ppf(alpha / 2, df)
chi2_critical_upper = chi2.ppf(1 - alpha / 2, df)

# Step 3: Interpret the results
print("Hypothesis Test for Population Variance:")
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Chi-square Statistic: {chi_square_stat:.2f}")
print(f"Critical Chi-square Values: [{chi2_critical_lower:.2f}, {chi2_critical_upper:.2f}]")

if chi_square_stat < chi2_critical_lower or chi_square_stat > chi2_critical_upper:
    print("Result: Reject the null hypothesis. The population variance is not equal to 25.")
else:
    print("Result: Fail to reject the null hypothesis. The population variance is equal to 25.")

# Step 4: Visualize the Chi-square distribution
x = np.linspace(0, chi2_critical_upper + 10, 500)
y = chi2.pdf(x, df)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f"Chi-square Distribution (df={df})", color="blue")
plt.axvline(chi2_critical_lower, color="red", linestyle="--", label=f"Critical Value Lower ({chi2_critical_lower:.2f})")
plt.axvline(chi2_critical_upper, color="green", linestyle="--", label=f"Critical Value Upper ({chi2_critical_upper:.2f})")
plt.axvline(chi_square_stat, color="orange", linestyle="-", label=f"Chi-square Stat ({chi_square_stat:.2f})")
plt.title("Chi-square Test for Variance", fontsize=16)
plt.xlabel("Chi-square Value", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.legend()
plt.grid(alpha=0.7)
plt.tight_layout()
plt.show()
```
